```python
# Bài 1
import numpy as np

# 1. Tạo một mảng NumPy chứa 20 giá trị ngẫu nhiên từ 0 đến 1
random_array = np.random.rand(20)
print("Mảng ngẫu nhiên từ 0 đến 1:", random_array)

# 2. Chuẩn hóa mảng này (đưa các giá trị về khoảng [0, 1])
normalized_array = (random_array - np.min(random_array)) / (np.max(random_array) - np.min(random_array))
print("Mảng sau khi chuẩn hóa:", normalized_array)

# 3. Tính tích vô hướng (dot product) của hai mảng 1D
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_product = np.dot(a, b)
print("Tích vô hướng của a và b:", dot_product)

# 4. Tạo một ma trận 5x5 và tính định thức, nghịch đảo
matrix = np.random.rand(5, 5)
print("Ma trận:", matrix)
determinant = np.linalg.det(matrix)
print("Định thức của ma trận:", determinant)

if determinant != 0:
    inverse_matrix = np.linalg.inv(matrix)
    print("Ma trận nghịch đảo:", inverse_matrix)
else:
    print("Ma trận không khả nghịch (định thức = 0).")
```

```python
import pandas as pd

# 1. Tạo DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [23, 25, 22, 24, 21],
    'Score': [85, 90, 78, 92, 88]
}
df = pd.DataFrame(data)
print("DataFrame:", df)

# 2. Tính giá trị trung bình của cột "Score"
average_score = df['Score'].mean()
print("Giá trị trung bình của cột 'Score':", average_score)

# 3. Lọc các hàng có "Score" lớn hơn 85
filtered_df = df[df['Score'] > 85]
print("Các hàng có Score > 85:", filtered_df)
```

```python
import pandas as pd

# Đọc file CSV (giả sử file Iris.csv đã được tải và lưu trên máy)
iris_df = pd.read_csv('Iris.csv')  # Đảm bảo cung cấp đường dẫn chính xác tới file Iris.csv

# Loại bỏ khoảng trắng hoặc ký tự đặc biệt khỏi tên các cột
iris_df.columns = iris_df.columns.str.strip()

# Hiển thị thông tin tổng quan về dữ liệu
print("Thông tin tổng quan về dữ liệu:")
print(iris_df.info())

# Mô tả dữ liệu: thống kê cơ bản
print("Mô tả dữ liệu:")
print(iris_df.describe())

# Tính toán cơ bản cho cột 'sepal_length'
print("Giá trị trung bình SepalLengthCm:", iris_df['SepalLengthCm'].mean())
print("Giá trị lớn nhất SepalLengthCm:", iris_df['SepalLengthCm'].max())
print("Giá trị nhỏ nhất SepalLengthCm:", iris_df['SepalLengthCm'].min())
```

```python
import pandas as pd

# Tạo DataFrame chứa dữ liệu thiếu
data_with_missing = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [23, None, 25, 24, 22],
    'City': ['New York', 'Boston', None, 'Chicago', 'Boston'],
    'Salary': [60000, 52000, None, 58000, None]
}

df_missing = pd.DataFrame(data_with_missing)
print("Dữ liệu ban đầu:")
print(df_missing)

# 1. Điền giá trị thiếu trong cột Age bằng giá trị trung bình
mean_age = df_missing['Age'].mean()
df_missing['Age'] = df_missing['Age'].fillna(mean_age)  # Sử dụng phương thức fillna không có inplace
print("\nDữ liệu sau khi điền giá trị thiếu trong cột Age:")
print(df_missing)

# 2. Xóa các hàng có nhiều hơn 1 giá trị thiếu
df_missing = df_missing.dropna(thresh=len(df_missing.columns) - 1)
print("\nDữ liệu sau khi xóa các hàng có nhiều hơn 1 giá trị thiếu:")
print(df_missing)

# 3. Điền giá trị thiếu trong cột Salary bằng 50000
df_missing['Salary'] = df_missing['Salary'].fillna(50000)  # Sử dụng phương thức fillna không có inplace
print("\nDữ liệu sau khi điền giá trị thiếu trong cột Salary bằng 50000:")
print(df_missing)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Biểu đồ đường biểu diễn hàm số y = x^2 trên khoảng [-10, 10]
x = np.linspace(-10, 10, 100)
y = x ** 2

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2')
plt.title('Biểu đồ đường của hàm số y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# 2. Biểu đồ cột thể hiện điểm số của các sinh viên từ Bài tập 3
# Dữ liệu từ Bài tập 3
data_scores = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Score': [85, 90, 78, 92, 88]
}
df_scores = pd.DataFrame(data_scores)

plt.figure(figsize=(8, 6))
plt.bar(df_scores['Name'], df_scores['Score'], color='skyblue')
plt.title('Biểu đồ cột điểm số của sinh viên')
plt.xlabel('Tên sinh viên')
plt.ylabel('Điểm số')
plt.show()

# 3. Biểu đồ tròn thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris
# Đọc file CSV (giả sử file Iris.csv đã được tải và lưu trên máy)
iris_df = pd.read_csv('Iris.csv')  # Đảm bảo cung cấp đường dẫn chính xác tới file Iris.csv
species_counts = iris_df['Species'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Biểu đồ tròn thể hiện phần trăm mỗi loại hoa trong tập dữ liệu Iris')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc file CSV (giả sử file Iris.csv đã được tải và lưu trên máy)
iris_df = pd.read_csv('Iris.csv')

# 1. Biểu đồ phân tán giữa sepal_length và sepal_width với màu sắc theo loại hoa
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=iris_df,
    x='SepalLengthCm',
    y='SepalWidthCm',
    hue='Species',
    palette='Set1',  # Sử dụng bảng màu phân loại để có màu sắc khác biệt
    s=100
)

# Thêm tiêu đề, nhãn trục và chú thích cho biểu đồ
plt.title('Biểu đồ phân tán Sepal Length và Sepal Width theo loại hoa')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.grid(True)

plt.show()

```
